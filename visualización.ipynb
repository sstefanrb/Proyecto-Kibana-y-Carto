```python
import osmnx as ox
import matplotlib.pyplot as plt
import pandas as pd
```

```python
ox.config(use_cache=True, log_console=True)
```

```python
lugar_de_interes = "Madrid, España"
etiqueta = {"shop": "toys"}

juguetes = ox.geometries_from_place(lugar_de_interes, tags=etiqueta)
```

```python
# Especifica las etiquetas para buscar colegios
tags = {"amenity": "school"}

# Obtiene las geometrías (ubicaciones) de gimnasios en el área especificada
colegios = ox.geometries_from_place(lugar_de_interes, tags)
```

```python
# Especifica las etiquetas para buscar guarderías
etiquetas_guarderias = {"amenity": "kindergarten"}

# Obtiene las geometrías (ubicaciones) de guarderías en el área especificada
guarderias = ox.geometries_from_place(lugar_de_interes, etiquetas_guarderias)
```

```python
# Visualiza las tiendas de juguetes y gimnasios en un mismo mapa
fig, ax = plt.subplots(figsize=(10, 10))
juguetes.plot(ax=ax, color="red", markersize=50, label="Tiendas de Juguetes")
colegios.plot(ax=ax, color="green", markersize=50, label="Colegios")
guarderias.plot(ax=ax, color="blue", markersize=30, label="Guarderias")
plt.legend()
plt.show()
```

```python
# Guarda las geometrías en un archivo para su uso posterior
juguetes.to_csv("tiendas_juguetes.csv")
colegios.to_csv("colegios.csv")
guarderias.to_csv("guarderias.csv")
```
